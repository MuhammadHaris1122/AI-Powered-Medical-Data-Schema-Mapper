# Step 3: Generate Column Metadata

In this step, we will create metadata for every column in both datasets.

The metadata includes:

- Column Name
- Data Type
- Number of Missing Values
- Number of Unique Values
- Sample Values

This metadata will later be used for AI-based schema matching.

```python
def generate_metadata(df):
    """
    Generate metadata for each column in a dataframe.
    """

    metadata = []

    for column in df.columns:

        metadata.append({
            "Column Name": column,
            "Data Type": str(df[column].dtype),
            "Missing Values": df[column].isnull().sum(),
            "Unique Values": df[column].nunique(),
            "Sample Values": list(df[column].dropna().unique()[:5])
        })

    metadata_df = pd.DataFrame(metadata)

    return metadata_df
```

# Step 6: Hybrid Schema Matching

In this step we improve the mapping algorithm by considering:

- Semantic Similarity
- Data Type Similarity
- ID Column Detection
- One-to-One Mapping

This reduces incorrect mappings and produces more reliable results.

```python
def is_id_column(column_name):
    """
    Check whether a column is an ID column.
    """

    column_name = column_name.lower()

    id_keywords = [
        "id",
        "record_id",
        "patient_id",
        "emp_id",
        "employee_id",
        "customer_id"
    ]

    return any(keyword in column_name for keyword in id_keywords)
```

Datatype Similarity

```python
def datatype_similarity(source_dtype, target_dtype):
    """
    Returns datatype similarity score.
    """

    if source_dtype == target_dtype:
        return 1.0

    return 0.0
```

Hybrid Score Function
similarity = 80 percent remains 

```python
def hybrid_similarity(
    semantic_score,
    source_col,
    target_col,
    source_dtype,
    target_dtype
):
    """
    Calculate final hybrid similarity score.
    """

    final_score = semantic_score * 0.80

    # Datatype bonus
    final_score += datatype_similarity(source_dtype, target_dtype) * 0.20

    # ID Rule
    source_is_id = is_id_column(source_col)
    target_is_id = is_id_column(target_col)

    if source_is_id != target_is_id:
        final_score = 0

    return final_score
```

```python
# Generate Hybrid Similarity Matrix
hybrid_matrix = similarity_matrix.copy()

for i, source_col in enumerate(source_columns):

    for j, target_col in enumerate(target_columns):

        hybrid_matrix[i][j] = hybrid_similarity(

            semantic_score=similarity_matrix[i][j],

            source_col=source_col,

            target_col=target_col,

            source_dtype=str(source_df[source_col].dtype),

            target_dtype=str(target_df[target_col].dtype)
        )
```

```python
hybrid_similarity_df = pd.DataFrame(

    hybrid_matrix,

    index=source_columns,

    columns=target_columns
)

display(hybrid_similarity_df)
```

Generate Final Mapping (One-to-One)

```python
mapping = []

used_targets = set()

for i, source_col in enumerate(source_columns):

    scores = hybrid_matrix[i].copy()

    while True:

        best_idx = scores.argmax()

        target_col = target_columns[best_idx]

        score = scores[best_idx]

        if target_col not in used_targets:

            used_targets.add(target_col)

            mapping.append({

                "Source Column": source_col,

                "Target Column": target_col,

                "Similarity Score": round(score,4)

            })

            break

        else:

            scores[best_idx] = -1
```

```python
mapping_df = pd.DataFrame(mapping)

display(mapping_df)
```

# Step 7: Medical Synonym Matching

In this step, we improve schema matching by adding a medical synonym dictionary.

Examples:
- Blood Sugar → Glucose
- BP → BloodPressure
- Sex → Gender
- Disease → Diagnosis

If two columns are known medical synonyms, their similarity score will be increased automatically.

```python
medical_synonyms = {

    "blood sugar": ["glucose", "blood_glucose"],

    "bp": ["bloodpressure", "blood_pressure"],

    "sex": ["gender"],

    "body mass index": ["bmi"],

    "disease": ["diagnosis"],

    "patient age": ["age"],

    "cholesterol level": ["cholesterol"],

    "serum insulin": ["insulin"],

    "record_id": ["patient_id"]
}
```

```python
# Synonym Matching Function
def synonym_similarity(source_col, target_col):
    """
    Returns 1 if source and target are medical synonyms.
    Otherwise returns 0.
    """

    source = source_col.lower().strip()
    target = target_col.lower().strip()

    # Exact match
    if source == target:
        return 1

    # Dictionary lookup
    if source in medical_synonyms:
        if target in medical_synonyms[source]:
            return 1

    return 0
```

# Update Hybrid Similarity Function

```python
def hybrid_similarity(
    semantic_score,
    source_col,
    target_col,
    source_dtype,
    target_dtype
):

    score = semantic_score * 0.70

    # Datatype Bonus
    score += datatype_similarity(source_dtype, target_dtype) * 0.20

    # Synonym Bonus
    score += synonym_similarity(source_col, target_col) * 0.10

    # ID Rule
    source_is_id = is_id_column(source_col)
    target_is_id = is_id_column(target_col)

    if source_is_id != target_is_id:
        return 0

    return score
```

```python
# regenrate hybrid matrix 
hybrid_matrix = similarity_matrix.copy()

for i, source_col in enumerate(source_columns):

    for j, target_col in enumerate(target_columns):

        hybrid_matrix[i][j] = hybrid_similarity(
            semantic_score=similarity_matrix[i][j],
            source_col=source_col,
            target_col=target_col,
            source_dtype=str(source_df[source_col].dtype),
            target_dtype=str(target_df[target_col].dtype)
        )
```

# generate final mapping  again

```python
mapping = []

used_targets = set()

for i, source_col in enumerate(source_columns):

    scores = hybrid_matrix[i].copy()

    while True:

        best_index = scores.argmax()

        target_col = target_columns[best_index]

        score = scores[best_index]

        if target_col not in used_targets:

            used_targets.add(target_col)

            mapping.append({
                "Source Column": source_col,
                "Target Column": target_col,
                "Similarity Score": round(score, 4)
            })

            break

        scores[best_index] = -1

mapping_df = pd.DataFrame(mapping)

display(mapping_df)
```

# Step 8: Multi-Level Matching Engine

In this step, we combine multiple matching techniques:

1. Exact Match
2. Medical Synonym Match
3. Fuzzy String Matching
4. Semantic Similarity
5. Data Type Validation

This produces a more accurate final mapping score.

```python
! pip install -q rapidfuzz
```

```python
from rapidfuzz import fuzz

def exact_match(source_col, target_col):
    """
    Returns 1 if both column names are exactly the same.
    """

    return source_col.lower().strip() == target_col.lower().strip()
```

```python
# fuzzt similarity
def fuzzy_similarity(source_col, target_col):
    """
    Calculate fuzzy similarity between two column names.
    """

    score = fuzz.token_sort_ratio(source_col, target_col)

    return score / 100
```

# improve hybrid score 

```python
def hybrid_similarity(
    semantic_score,
    source_col,
    target_col,
    source_dtype,
    target_dtype
):
    """
    Multi-Level Matching Score
    """

    # Rule 1: ID columns
    if is_id_column(source_col) != is_id_column(target_col):
        return 0

    # Rule 2: Exact Match
    if exact_match(source_col, target_col):
        return 1.0

    # Rule 3: Semantic Score
    score = semantic_score

    # Rule 4: Datatype Bonus
    if datatype_similarity(source_dtype, target_dtype):
        score += 0.10

    # Rule 5: Synonym Bonus
    if synonym_similarity(source_col, target_col):
        score += 0.15

    # Rule 6: Fuzzy Bonus
    fuzzy_score = fuzzy_similarity(source_col, target_col)

    if fuzzy_score >= 0.80:
        score += 0.10
    elif fuzzy_score >= 0.60:
        score += 0.05

    # Final score cannot exceed 1
    score = min(score, 1.0)

    return round(score, 4)
```

# regenrate hybrid matrix

```python
hybrid_matrix = similarity_matrix.copy()

for i, source_col in enumerate(source_columns):

    for j, target_col in enumerate(target_columns):

        hybrid_matrix[i][j] = hybrid_similarity(
            semantic_score=similarity_matrix[i][j],
            source_col=source_col,
            target_col=target_col,
            source_dtype=str(source_df[source_col].dtype),
            target_dtype=str(target_df[target_col].dtype)
        )
```

```python
# generate mapping
mapping = []
used_targets = set()

for i, source_col in enumerate(source_columns):

    scores = hybrid_matrix[i].copy()

    while True:

        best_index = scores.argmax()

        target_col = target_columns[best_index]

        if target_col not in used_targets:

            mapping.append({
                "Source Column": source_col,
                "Target Column": target_col,
                "Similarity Score": round(scores[best_index], 4)
            })

            used_targets.add(target_col)
            break

        scores[best_index] = -1

mapping_df = pd.DataFrame(mapping)

display(mapping_df)
```
